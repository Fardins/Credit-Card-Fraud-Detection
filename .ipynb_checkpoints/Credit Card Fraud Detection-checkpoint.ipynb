#### <div style="color: green;">Dataset description:</div>
- **Number of Entries:** 284,807
- **Number of Columns:** 31

- **Columns:**
  - **Time:** The time elapsed between this transaction and the first transaction in the dataset (in seconds).
  - **V1 to V28:** These are the result of a PCA transformation; their meanings are not directly interpretable, as they are anonymized features.
  - **Amount:** The transaction amount.
  - **Class:** The target variable, where 1 indicates a fraudulent transaction and 0 indicates a legitimate transaction.

#### <div style="color: green;">Key Features:</div>

In this dataset, some key features that are typically considered important for credit card fraud detection include:

- **Time:** This feature represents the elapsed time between each transaction and the first transaction in the dataset. It might capture temporal patterns in fraudulent activities.

- **V1 to V28:** These features are the result of a Principal Component Analysis (PCA), a dimensionality reduction technique applied to anonymize the original features. Although their exact meanings are not interpretable, they are engineered to capture important variance in the data. Some of these components may be more strongly associated with fraud.

- **Amount:** The transaction amount can be a crucial indicator, as fraudulent transactions might involve unusually high or low amounts.

- **Class:** This is the target variable indicating whether a transaction is fraudulent (1) or not (0). The relationship between the other features and this class label is what the model will learn to predict fraud.

The PCA-transformed features (V1 to V28) are particularly important because they encapsulate the most significant variance in the original data. Typically, during feature importance analysis, some of these components may be found to have higher importance than others. Additionally, the Amount feature often shows significant importance since fraud might correlate with abnormal transaction amounts.

#### <div style="color: green;">Target Variable:</div>

The target variable in this dataset is the **"Class"** column. It indicates whether a transaction is fraudulent or not:

- **0:** Indicates a legitimate transaction.
- **1:** Indicates a fraudulent transaction.

This binary classification target is used to train models to detect fraud based on the other features in the dataset.

#### <div style="color: green;"> Handle Outliers:</div>

```python
#to see outlier in data
data_numeric_cols = [col for col in data.columns if data[col].dtype in ['int64', 'float64']]

plt.figure(figsize=(15,15))
for i, col in enumerate(data_numeric_cols):
    plt.subplot(6, 6, i+1)
    sns.boxplot(x='Class', y=col, data=data)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Handle Outlier

data_num = data.select_dtypes(include='number')

for column in data_num.columns:
    for i in data["Class"].unique():
        selected_i = data[data["Class"] == i]
        selected_column = selected_i[column]

        Q1 = selected_column.quantile(0.25)
        Q3 = selected_column.quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = selected_column[(selected_column < lower_bound) | (selected_column > upper_bound)].index
        data.drop(index=outliers, inplace=True)
        print(column, i, outliers)
```

#### <div style="color: green;"> Histogram Analysis:</div>

```python
def draw_histograms(dataframe, features, rows, cols):
    colors = ['tan']
    
    fig = plt.figure(figsize=(20, 20))
    for i, feature in enumerate(features):
        ax = fig.add_subplot(rows, cols, i + 1)
        dataframe[feature].hist(bins=20, ax=ax, facecolor=colors[i % len(colors)])
        ax.set_title(feature + " Distribution", color='DarkRed')
        ax.set_yscale('log')  # Set y-axis to a logarithmic scale for better visibility
    fig.tight_layout()  
    plt.show()

# Plot histograms for all columns in the dataset in an 8x4 grid with different colors for each plot
draw_histograms(data, data.columns, 8, 4)
```

#### Observations:
From Histogam analysis we see that-
- **Imbalanced Dataset:** The most critical observation is the significant imbalance in the **Class** variable, indicating that most transactions are legitimate. This imbalance needs to be addressed when developing models to prevent bias towards non-fraudulent predictions.
- **Skewness in Transaction Amounts:** The Amount variable is skewed, and this could be a key feature in distinguishing between fraudulent and non-fraudulent transactions.
- **Uniform Distribution of Time:** The even distribution of Time suggests that fraud is not time-dependent in this dataset.

### <div style="color: gray;">Step 3:Model Training and Performance Evaluation-</div> 

#### <div style="color: green;"> Split the dataset into training and testing sets:</div>


```python
# Separate properties and target variable

X = data.drop("Class", axis=1)
Y = data.Class
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state = 4, test_size = 0.2)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

#### <div style="color: green;"> Normalizing the data:</div>

```python
from sklearn.preprocessing import StandardScaler

s = StandardScaler()
x_train = s.fit_transform(x_train)
x_test = s.fit_transform(x_test)
```

#### <div style="color: green;"> All Regression Model Training and Evaluation:</div>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
svc = SVC()
gnb = GaussianNB()
dtc = DecisionTreeClassifier(max_depth=5)
lrc = LogisticRegression(solver='liblinear', penalty='l1')
rfc = RandomForestClassifier(n_estimators=50, random_state=2)
abc = AdaBoostClassifier(n_estimators=50, random_state=2)
bc = BaggingClassifier(n_estimators=50, random_state=2)
etc = ExtraTreesClassifier(n_estimators=50, random_state=2)
gbdt = GradientBoostingClassifier(n_estimators=50,random_state=2)
xgb = XGBClassifier(n_estimators=50,random_state=2)
```

```python
clfs = {
    'SVC' : svc,
    'NB': gnb,
    'DT': dtc, 
    'LR': lrc, 
    'RF': rfc, 
    'AdaBoost': abc, 
    'BgC': bc, 
    'ETC': etc,
    'GBDT':gbdt,
    'xgb':xgb
}
```

```python
def train_classifier(clf, x_train, y_train, x_test, y_test):
    try:
        clf.fit(x_train, y_train)
        y_pred = clf.predict(x_test)
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        return accuracy, precision, recall, f1
    except Exception as e:
        print(f"Error with classifier {clf.__class__.__name__}: {str(e)}")
        return None, None
```

#### <div style="color: green;">Evaluate Performence:</div>


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,RocCurveDisplay
from termcolor import colored
```

```python
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for name,clf in clfs.items():
    current_accuracy,current_precision,current_recall,current_f1 = train_classifier(clf, x_train,y_train,x_test,y_test)
    
    print("For ",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)
    print("Recall - ",current_recall)
    print("F1 - ",current_f1)
    
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    recall_scores.append(current_recall)
    f1_scores.append(current_f1)
```

```python
performance_data = pd.DataFrame({'Algorithm':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores, 'Recall':recall_scores, 'F1':f1_scores})
```

```python
performance_data
```

```python
performance_data.sort_values('Accuracy',ascending=False)
```

#### <div style="color: green;">Performance Observation:</div>

- **Overall Performance:**
  - All algorithms are performing exceptionally well, with accuracy scores above 99.5%. 
  - This could indicate either a very well-separated dataset or potential overfitting.

- **Best Performers:**
  - SVC, DT (Decision Tree), LR (Logistic Regression), RF (Random Forest), BgC (Bagging Classifier), ETC (Extra Trees Classifier), and XGBoost all achieved the same top scores across all metrics.

- **Precision:**
  - Most algorithms achieved perfect precision (1.0), except AdaBoost (0.9459) and GBDT (0.875).

- **Recall:**
  - Recall is consistently 0.875 for all algorithms except NB (Naive Bayes), which has a slightly lower recall of 0.85.

- **F1 Score:**
  - The F1 score, which is the harmonic mean of precision and recall, is highest (0.9333) for the top performers and lowest for GBDT (0.875).

- **Slight Variations:**
  - NB has slightly lower recall and F1 score compared to the top performers.
  - AdaBoost has slightly lower precision and F1 score.
  - GBDT has the lowest precision and F1 score among all, but still performs very well.

### <div style="color: gray;">Step 4: Hyperparameter Tuning-</div>

- I use RandomizedSearchCV for hyperparameter optimization. It basically works with various parameters internally and finds out the best parameters.
- I apply Hyperparameter tuning technique in our SVC and LR (Logistic Regression) Model because these four model gives present higher accuracy.

#### <div style="color: green;">Initialized Hyperparameters:</div>

```python
#import hyperparameter
from sklearn.model_selection import RandomizedSearchCV
```

```python
#I use RandomizedSearchCV for hyperparameter optimization

from scipy.stats import randint, uniform, loguniform

#Define hyperparameters for Support Vector Machine (SVM)
svm_params = {
    'C': uniform(loc=0.1, scale=10),         
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  
    'degree': np.arange(2, 5),               
    'gamma': ['scale', 'auto', uniform(loc=0.01, scale=0.1)], 
    'coef0': uniform(0, 10),        
    'class_weight': [None, 'balanced']      
}

#Define hyperparameters for Logistic Regression
lrc_params = {
    'C': uniform(loc=0.001, scale=10),  # Regularization strength
    'solver': ['liblinear', 'lbfgs', 'saga'],  # Optimization algorithm
    'penalty': ['l1', 'l2', 'elasticnet', None],  # Type of regularization
    'max_iter': [100, 200, 300, 400, 500],
    'l1_ratio': np.linspace(0, 1, 10)  # Only used if penalty is elasticnet
}
```

#### <div style="color: green;">Support Vector Machine (SVM) classifier Model for DRandomizedSearchCV Hyperparameter:</div>


```python
svc = SVC()

svm_random_search = RandomizedSearchCV(
    svc, param_distributions=svm_params, n_iter=100, cv=5, verbose=2, random_state=42, n_jobs=-1
)
```

```python
svm_random_search.fit(x_train, y_train)
```

```python
svm_random_search.best_params_
```

#### <div style="color: green;">Evaluate Support Vector Machine (SVM) Classifier Model for DRandomizedSearchCV Hyperparameter:</div>


```python
y_pred= svm_random_search.predict(x_test)

train_accuracy_SVC_RSH = svm_random_search.score(x_train,y_train)
test_accuracy_SVC_RSH = svm_random_search.score(x_test, y_test)
accuracy_score_SVC_RSH = accuracy_score(y_test, y_pred)
precision_score_SVC_RSH = precision_score(y_test, y_pred)
recall_score_SVC_RSH = recall_score(y_test, y_pred)
f1_score_SVC_RSH = f1_score(y_test, y_pred)
cm_SVC_RSH = confusion_matrix(y_test, y_pred)

print(colored('Support Vector Machine (SVM) with DRandomizedSearchCV Hyperparameter Model Evaluation:\n',color = 'blue', attrs = ['bold','dark']))
print(colored(f'train_accuracy : {round(train_accuracy_SVC_RSH,6)}',color='light_magenta'))
print(colored(f'test_accuracy : {round(test_accuracy_SVC_RSH,6)}',color='light_magenta'))
print(colored(f'accuracy_score : {round(accuracy_score_SVC_RSH,6)}',color='light_magenta'))
print(colored(f'precision_score : {round(precision_score_SVC_RSH,6)}',color='light_magenta'))
print(colored(f'recall_score : {round(recall_score_SVC_RSH,6)}',color='light_magenta'))
print(colored(f'f1_score : {round(f1_score_SVC_RSH,6)}',color='light_magenta'))
print(colored(f'\nconfusion_matrix :\n {cm_SVC_RSH}',color='light_yellow'))
```

#### <div style="color: green;"> Logistic Regression Model for DRandomizedSearchCV Hyperparameter:</div>


```python
lrc = LogisticRegression()

lrc_random_search = RandomizedSearchCV(
    estimator=lrc,
    param_distributions=lrc_params,
    n_iter=100,  # Number of parameter settings sampled
    cv=5,  # Number of cross-validation folds
    verbose=2,
    random_state=42,
    n_jobs=-1  # Use all available cores
)
```

```python
lrc_random_search.fit(x_train, y_train)
```

```python
lrc_random_search.best_params_
```

#### <div style="color: green;">Evaluate Logistic Regression Model for DRandomizedSearchCV Hyperparameter:</div>

```python
y_pred= lrc_random_search.predict(x_test)

train_accuracy_LR_RSH = lrc_random_search.score(x_train,y_train)
test_accuracy_LR_RSH = lrc_random_search.score(x_test, y_test)
accuracy_score_LR_RSH = accuracy_score(y_test, y_pred)
precision_score_LR_RSH = precision_score(y_test, y_pred)
recall_score_LR_RSH = recall_score(y_test, y_pred)
f1_score_LR_RSH = f1_score(y_test, y_pred)
cm_LR_RSH = confusion_matrix(y_test, y_pred)

print(colored('Logistic Regression with DRandomizedSearchCV Hyperparameter Model Evaluation:\n',color = 'blue', attrs = ['bold','dark']))
print(colored(f'train_accuracy : {round(train_accuracy_LR_RSH,6)}',color='light_magenta'))
print(colored(f'test_accuracy : {round(test_accuracy_LR_RSH,6)}',color='light_magenta'))
print(colored(f'accuracy_score : {round(accuracy_score_LR_RSH,6)}',color='light_magenta'))
print(colored(f'precision_score : {round(precision_score_LR_RSH,6)}',color='light_magenta'))
print(colored(f'recall_score : {round(recall_score_LR_RSH,6)}',color='light_magenta'))
print(colored(f'f1_score : {round(f1_score_LR_RSH,6)}',color='light_magenta'))
print(colored(f'\nconfusion_matrix :\n {cm_LR_RSH}',color='light_yellow'))
```

- **F1 Score:**
